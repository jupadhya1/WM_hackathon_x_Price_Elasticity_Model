```python
df_cust_seg = (df_cust_seg.drop(columns='member_number_nunique')
                        .rename(columns={'member_number_first':'member_number',
                                        'wholesale_flag_first': 'wholesale_flag',
                                        'card_no':'MEMBER_ID'}))
```

# <span style="color:#c6aa3d">5. Substitution analysis</span><a class="anchor" id="5-bullet"></a>

### 5.1 Preparing data for substitution

```python
# Pulling config parameters
lvl = config.subst_lvl
min_customer_support = config.subst_min_customer_support # Increase support to filter out brand pairs by making overlap criterion more strict
store_banner = config.subst_store_banner # Run twice, once for each store
```

```python
# Set customer identifier- default= member_number
subst_df['cust_id'] = subst_df[config.subst_customer_identifier]
```

```python
# Store banner filtering
df = subst_df.loc[subst_df['banner']==store_banner]
```

**Note**: In the below code, the term `sku` has been used only to imply the level at which we are conducting the analysis. The level is given by the `subst_lvl` parameter in the config file

### 5.2 Preparing customer yule metric

#### 5.2.1 Cross-join and calculate intersection

```python
# Cross-tabulate
df_customer = df[['SUBCLASS','cust_id',lvl]].drop_duplicates()
cross_skus = df_customer.merge(df_customer.add_prefix('right_'), left_on=[
                              'SUBCLASS','cust_id'], right_on=['right_SUBCLASS','right_cust_id'])

# Create intersection
customer_intersect = (cross_skus
             .groupby(['SUBCLASS', lvl, 'right_'+lvl])
             .size()
             .reset_index()
             .rename(columns={0: 'sku_intersect_count'})
             )
```

#### 5.2.2 Calculate other metrics for yule score calculation and merge

```python
# Create the SKU summary: number of customers that bought SKU
cust_per_sku = (df_customer
                 .groupby(['SUBCLASS', lvl])
                 .agg({'cust_id': pd.Series.nunique})
                 .reset_index()
                 .rename(columns={'cust_id': 'sku_count'})
                 )

# Create the total number of customers
customer_total = df_customer.groupby('SUBCLASS')['cust_id'].nunique().reset_index()
customer_total = customer_total.rename(columns={'cust_id': 'total_count'})

# Merge the data from SKU A, SKU B, and totals
yule_customer = (customer_intersect
                .merge(cust_per_sku, on=['SUBCLASS', lvl], how='inner')
                .rename(columns={'sku_count': 'left_sku_count'})
                .merge(cust_per_sku.rename(columns={lvl: 'right_'+lvl}), on=['SUBCLASS', 'right_'+lvl], how='inner')
                .rename(columns={'sku_count': 'right_sku_count'})
                .merge(customer_total, on='SUBCLASS', how='inner')
                )

# Calculate support for intersection
yule_customer['support_customer'] = yule_customer['sku_intersect_count']/yule_customer['total_count']
```

#### 5.2.3 Calculate customer yule score and filter same sku and low support pairs

```python
# Calculate yule values and odd ratios from the metrics at customer level for each pair of SKUs
yule_customer = (yule_customer.assign
                  (
                      # Odds left/right and ratio
                      l_odds=lambda x: x.sku_intersect_count / \
                      (x.left_sku_count - x.sku_intersect_count),
                      r_odds=lambda x: (x.right_sku_count - x.sku_intersect_count) / \
                          (x.total_count - x.left_sku_count - x.right_sku_count + x.sku_intersect_count),
                      odds_ratio=lambda x: x.l_odds / x.r_odds,

                      # Yules metric
                      yulesQ=lambda x: (x.odds_ratio - 1) / (x.odds_ratio+1)
                    )
                  .rename(columns={'sku_intersect_count': 'overlap_customer_level', 
                                   'left_sku_count': 'freq_customer_level', 
                                   'right_sku_count': 'right_freq_customer_level', 
                                   'total_count': 'total_count_customers'})
                )
                  
# Remove rows with same SKUs
yule_customer = yule_customer[~(yule_customer[lvl]==yule_customer['right_'+lvl])]

# Remove combinations with very less support
yule_customer = yule_customer[yule_customer['support_customer']>=min_customer_support]
```

### 5.3 Preparing transaction yule metric

```python
# Create transaction ID
df['transaction_id'] = df['Date'].astype(str) + '_' + df['STORE_NUMBER'].astype(str) + \
                        '_' + df['POS_NUMBER'].astype(str) + '_' + df['TICKET_NUMBER'].astype(str)
```

#### 5.3.1 Cross-join and calculate intersection

```python
## YulesQ at transaction level
df_transaction = df[['SUBCLASS','transaction_id', lvl]].drop_duplicates()

cross_transactions = (df_transaction
                      .merge(df_transaction.add_prefix('right_'), left_on=['SUBCLASS','transaction_id'], 
                             right_on=['right_SUBCLASS','right_transaction_id'], how='inner')
                      .drop(columns=['right_transaction_id'])
                      )

transaction_intersect = (cross_transactions
             .groupby(['SUBCLASS', lvl, 'right_'+lvl])
             .size()
             .reset_index()
             .rename(columns={0: 'sku_intersect_count'})
             )
```

#### 5.3.2 Calculate other metrics for yule score calculation and merge

```python
transactions_per_skus = (df_transaction
                         .groupby(['SUBCLASS', lvl])
                         .agg({'transaction_id': pd.Series.nunique})
                         .reset_index()
                         .rename(columns={'transaction_id': 'sku_count'})
                         )

transactions_total = df_transaction.groupby('SUBCLASS')['transaction_id'].nunique().reset_index()
transactions_total = transactions_total.rename(columns={'transaction_id': 'total_count'})

# Merge the three data sets to get the numbers needed to compute yule
yule_transaction = (transaction_intersect
                    .merge(transactions_per_skus, on=['SUBCLASS', lvl], how='inner')
                    .rename(columns={'sku_count': 'left_sku_count'})
                    .merge(transactions_per_skus.rename(columns={lvl: 'right_'+lvl}), 
                           on=['SUBCLASS', 'right_'+lvl], how='inner')
                    .rename(columns={'sku_count': 'right_sku_count'})
                    .merge(transactions_total, on='SUBCLASS', how='inner')
                    )

yule_transaction['support_transaction'] = yule_transaction['sku_intersect_count']/yule_transaction['total_count']
```

#### 5.3.3 Calculate transaction/basket yule score

```python
# Calculate yule values and odd ratios (same as before)
yule_transaction = (yule_transaction.assign
                    (
                        l_odds=lambda x: x.sku_intersect_count /
                        (x.left_sku_count - x.sku_intersect_count),
                        r_odds=lambda x: (x.right_sku_count - x.sku_intersect_count) / (
                            x.total_count - x.left_sku_count - x.right_sku_count + x.sku_intersect_count),
                        odds_ratio=lambda x: x.l_odds / x.r_odds,
                        yulesQ=lambda x: (x.odds_ratio - 1) / (x.odds_ratio+1)
                    )
                    .rename(columns={'sku_intersect_count': 'overlap_transaction_level', 
                                     'left_sku_count': 'freq_transaction_level', 
                                     'right_sku_count': 'right_freq_transaction_level', 
                                     'total_count': 'total_count_transaction'})
                   )
```

### 5.4 Calculate overall yule score

```python
# Merge transaction yule with customer yule results
yule = yule_customer.merge(yule_transaction, on=['SUBCLASS' ,lvl, 'right_' + lvl], 
                           how='left', suffixes=('_customer','_transaction'))

# Fill NAs with -1 as these are pairs with no intersection at transaction level
yule.yulesQ_transaction = yule.yulesQ_transaction.fillna(-1)

# Compute overall yules score
yule['substitution_yules'] = yule.apply(lambda x: (
        x.yulesQ_customer - x.yulesQ_transaction) if (x.yulesQ_customer >= x.yulesQ_transaction) else 0, axis=1)

# Compute standardized score which scales values typically from 0 to 1
yule['substitution_yules_std'] = yule['substitution_yules'] / 2

# Sort based on highest substitutable pair within subclass and brand
yule.sort_values(['SUBCLASS', lvl,'substitution_yules'], ascending = [True, True, False], inplace = True)

# Arranging columns
yule = yule[['SUBCLASS', lvl, 'right_'+lvl,
        'support_customer','support_transaction',
        'overlap_customer_level','freq_customer_level','right_freq_customer_level',
        'overlap_transaction_level','freq_transaction_level','right_freq_transaction_level',
        'l_odds_customer','r_odds_customer','odds_ratio_customer','yulesQ_customer',
        'l_odds_transaction','r_odds_transaction','odds_ratio_transaction','yulesQ_transaction',
        'substitution_yules','substitution_yules_std'
            ]]

yule = yule.rename(columns=config.subst_output_rename_dict)

print(f"Final output shape: {yule.shape}")

yule.head()
```

### 5.5 Exporting results

```python
# Export to csv
yule.to_csv(notebook_output_path + 'Substitutability_output_'+store_banner+'.csv',
                                                       index=False,
                                                       encoding='utf_8_sig')
```
